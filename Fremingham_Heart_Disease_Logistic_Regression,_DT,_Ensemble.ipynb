```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Calculate the number of missing values
df.isna().apply(pd.value_counts).T
```

```python
# Replace the null value with mode 
df['education'].fillna(df['education'].mode()[0],inplace=True)
```

```python
df['cigsPerDay'].fillna(df['cigsPerDay'].mode()[0],inplace=True)



```

```python
df['BPMeds'].fillna(df['BPMeds'].mode()[0],inplace=True)

```

```python
df['totChol'].fillna(df['totChol'].mode()[0],inplace=True)


```

```python
df['BMI'].fillna(df['BMI'].mode()[0],inplace=True)

```

```python

df['heartRate'].fillna(df['heartRate'].mode()[0],inplace=True)
```

```python
df['glucose'].fillna(df['glucose'].mode()[0],inplace=True)

```

```python
# Recheck for missing values
df.isna().apply(pd.value_counts).T
```

```python
df.info()
```

```python
# Selection of Variables 
X=df.drop(['age','currentSmoker','diabetes','prevalentStroke'],axis=1)
```

```python
y=df[['prevalentStroke']]
```

```python
# Scaling the Data using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
s=MinMaxScaler()
```

```python
x_scaled = s.fit_transform(X)
```

```python
X=pd.DataFrame(x_scaled,columns=X.columns)
```

```python
X.head()
```

```python
#splitting the data 
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=70,stratify=y)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train
```

```python
y_train
```

```python
#importing of Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
model.fit(X_train,y_train)
```

```python
DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# Calculate accuracy 
accuracy_score(y_test,y_pred)*100
```

```python
# Check model score -test
model.score(X_test,y_test)*100
```

```python
# check model score -train
model.score(X_train,y_train)*100
```
