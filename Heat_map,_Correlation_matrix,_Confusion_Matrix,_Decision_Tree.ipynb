```python
import pandas as pd
import numpy as np


df=pd.read_csv('/content/ADANIPORTS.csv')
```

```python
df.head()
```

```python
# Calculate missing values 
df.isna().apply(pd.value_counts).T
```

```python
df['Trades']=df['Trades'].replace('np.nan')
```

```python
df.isna().apply(pd.value_counts).T
```

```python
df['Trades'].fillna(df['Trades'].mode()[0],inplace=True)
```

```python
df.isna().apply(pd.value_counts).T
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Lebel Encoding 
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['Symbol']=le.fit_transform(df['Symbol'])
```

```python
df['Date'] = le.fit_transform(df['Date'])
```

```python
df['Series'] = le.fit_transform(df['Series'])
```

```python
df.info()
```

```python
df.head()
```

```python
# Data Selection 
X=df.drop(['Date','Symbol'],axis=1)
```

```python
y=df[['Symbol']]
```

```python
# Scaling the Data using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
s=MinMaxScaler()
```

```python
x_scaled = s.fit_transform(X)
```

```python
X=pd.DataFrame(x_scaled,columns=X.columns)
```

```python
X.head()
```

```python
#splitting the data 
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=50,stratify=y)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train
```

```python
y_train
```

```python
#importing of Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
model.fit(X_train,y_train)
```

```python
DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# Calculate accuracy 
accuracy_score(y_test,y_pred)*100
```

```python
# Check model score 
model.score(X_test,y_test)*100
```

```python
model.score(X_train,y_train)*100
```
